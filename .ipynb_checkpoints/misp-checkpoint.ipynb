```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import sys
#sys.path.append('../..')  # path to pymhlib to be used
sys.path.append("C:/Users/Desktop/pymhlib")
%load_ext autoreload
%aimport pymhlib.demos.misp, logging
%autoreload 1
from pymhlib.settings import settings, parse_settings
from pymhlib.log import init_logger
from pymhlib.demos.misp import MISPInstance, MISPSolution
from pymhlib.gvns import GVNS, Method
from pymhlib.solution import Solution, TObj
from typing import Callable, List, Any, Optional
import logging
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import networkx as nx
import time
import math
sns.set()
#%matplotlib inline
#from IPython.display import set_matplotlib_formats; set_matplotlib_formats('png', 'pdf')
if not settings.__dict__: parse_settings(args='')
init_logger()
logger = logging.getLogger("pymhlib")
#runs=3
problem_name = "gnm-20-50"
settings.mh_titer=200
settings.mh_lfreq = 1
settings.mh_tciter = 50
settings.seed = 13

logger.info(f"pymhlib demo for solving {problem_name}")
instance = MISPInstance(problem_name)
graph = instance.graph





logger.info(f"{problem_name} instance read:\n" + str(instance))
solution = MISPSolution(instance)

class MyGVNS(GVNS):
    
        sol_iter = dict()
    
        def __init__(self, sol: Solution, meths_ch: List[Method], meths_li: List[Method], meths_sh: List[Method], own_settings: dict = None, consider_initial_sol=False):
            super().__init__(sol, meths_ch, meths_li, meths_sh, own_settings, consider_initial_sol)


        def log_iteration(self, method_name: str, obj_old: TObj, new_sol: Solution, new_incumbent: bool, in_any_case: bool,log_info: Optional[str]):
            log = in_any_case or new_incumbent and self.own_settings.mh_lnewinc
            self.sol_iter[self.iteration] = new_sol.x[:new_sol.sel]
            if not log:
                lfreq = self.own_settings.mh_lfreq
                if lfreq > 0 and self.iteration % lfreq == 0:
                    log = True
                elif lfreq < 0 and self.is_logarithmic_number(self.iteration):
                    log = True
            if log:
                s = f"I {self.iteration:>10d} {self.incumbent.obj():16.5f} {obj_old:16.6f} {new_sol.obj():16.5f} "

                if self.population is not None:
                    s += f"{self.population.obj_avg():16.6f} {self.population.obj_std():16.5f} "

                s += f"{time.process_time()-self.time_start:12.4f} " \
                    f"{method_name:<20} {log_info if log_info is not None else ''}"
                self.sol_iter[self.iteration] = (method_name, new_incumbent, new_sol.x[:new_sol.sel], self.incumbent.x[:self.incumbent.sel])
                self.iter_logger.info(s)
```

```python
    
alg = MyGVNS(solution,
            [Method(f"ch0", MISPSolution.construct, 1)],
            [Method(f"li2", MISPSolution.local_improve, 2)],
            [Method(f"sh{i}", MISPSolution.shaking, i) for i in range(1,10)])



alg.run()

alg.method_statistics()
alg.main_results()
#for i,v in alg.sol_iter.items():
#    logger.info(alg.sol_iter[i])
sol = solution.x[:solution.sel]
col = ['r' if i in sol else 'b' for i in range(len(graph.nodes()))]

#nx.draw(graph, pos, with_labels=True, node_color=col)



```
